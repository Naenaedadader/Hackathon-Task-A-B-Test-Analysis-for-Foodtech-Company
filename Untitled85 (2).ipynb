# 4.2 執行 T 檢定&卡方 (4.2 Perform T-Test & Chi-Square)

```python
# 修正 T-Test 函式 (互換順序) (Fixed T-Test function (swap order))
def perform_t_test(df, group_name):
    t_stat, p_value = ttest_ind(df[df["variation"] == 2]["converted"],  # 互換順序
                    df[df["variation"] == 1]["converted"], equal_var=False)
    print(f"\n {group_name} - T-Test Result: T-Statistic = {t_stat:.4f}, P-Value = {p_value:.8f}")

# 修正 Chi-Square Test 函式 (Modified Chi-Square Test function)
def perform_chi2_test(df, group_name):
    conversion_table = pd.crosstab(df["variation"], df["converted"])
    chi2_stat, p_value, _, _ = chi2_contingency(conversion_table)
    print(f"\n {group_name} - Chi-Square Test Result: Chi-Square Statistic = {chi2_stat:.4f}, P-Value = {p_value:.8f}")

# 執行測試 (Executing the test)
perform_t_test(df_1, "df_1 (Original Data)")
perform_t_test(df_2, "df_2 (Drop final_order_status NaN)")
perform_t_test(df_3, "df_3 (Drop all NaN)")

perform_chi2_test(df_1, "df_1 (Original Data)")
perform_chi2_test(df_2, "df_2 (Drop final_order_status NaN)")
perform_chi2_test(df_3, "df_3 (Drop all NaN)")
```

```python
def perform_general_chi2_test(df, group_name, col_name):
    """
    針對 A/B 測試數據，對指定的分類變數與 converted 進行卡方檢定
    :param df: 數據框
    :param group_name: 數據集名稱
    :param col_name: 要檢測的分類變數（如 variation, platform, event_type）
    """
    #For A/B test data, perform a chi-square test on the specified categorical variable and converted
    #:param df: data frame
    #:param group_name: dataset name
    #:param col_name: the categorical variable to be tested (such as variation, platform, event_type)
    conversion_table = pd.crosstab(df[col_name], df["converted"])
    chi2_stat, p_value, _, _ = chi2_contingency(conversion_table)

    print(f"\n {group_name} - Chi-Square Test ({col_name} vs. converted):")
    print(f"  Chi-Square Statistic = {chi2_stat:.4f}, P-Value = {p_value:.8f}")

# 🚀 執行卡方檢定
for df_name, df in zip(["df_1 (Original Data)", "df_2 (Drop shop_id NaN)", "df_3 (Drop final_order_status NaN)"], [df_1, df_2, df_3]):
    perform_general_chi2_test(df, df_name, "variation")  # A/B 測試組別 vs. 轉化
    perform_general_chi2_test(df, df_name, "platform")   # 平台 (iOS/Android) vs. 轉化
    perform_general_chi2_test(df, df_name, "event_type") # 事件類型 vs. 轉化
```

```python
# 複製 df_3 進行處理 (Copy df_3 for processing)
print("\nStep 1: 處理缺失值，插補 NaN") #\nStep 1: Handle missing values ​​and interpolate NaN
df_imputed = df_3.copy()
df_imputed["final_order_status"].fillna("NotPaid", inplace=True)
df_imputed["converted"] = (df_imputed["final_order_status"] == "successful").astype(int)

# 分層抽樣 (Stratified sampling)
print("\nStep 2: 進行分層抽樣 (Stratified Sampling)") #\nStep 2: Stratified Sampling
df_stratified = df_imputed.groupby("event_type", group_keys=False).apply(lambda x: x.sample(frac=0.5, random_state=42))

# 轉化率比較 (Conversion rate comparison)
conversion_rates = pd.DataFrame({
    "df_3": df_3.groupby("variation")["converted"].mean(),
    "Imputed NaN": df_imputed.groupby("variation")["converted"].mean(),
    "Stratified Sampling": df_stratified.groupby("variation")["converted"].mean(),
})
print("\nA/B Test Conversion Rate Comparison:")
print(conversion_rates)

# 定義不同的 frac 值來測試. (Define different frac values ​​to test)
frac_values = [0.3, 0.5, 0.7]
results = []

# 插補 NaN (Imputation) A/B 測試結果. (Imputation of NaN A/B test results)
df_imputed_conversion_rate = df_imputed.groupby("variation")["converted"].mean()

# 進行 T 檢定 (Imputation). Perform T test (Imputation).
group_1_imputed = df_imputed[df_imputed["variation"] == 1]["converted"]
group_2_imputed = df_imputed[df_imputed["variation"] == 2]["converted"]
t_stat_imputed, p_value_t_imputed = ttest_ind(group_2_imputed, group_1_imputed, equal_var=False)

# 進行卡方檢定 (Imputation). Perform a Chi-square test (Imputation).
conversion_table_imputed = pd.crosstab(df_imputed["variation"], df_imputed["converted"])
chi2_stat_imputed, p_value_chi_imputed, _, _ = chi2_contingency(conversion_table_imputed)

# 存儲插補 NaN (Imputation) 的結果. (Storing the results of imputation of NaNs)
results.append({
    "method": "Imputation",
    "frac": "N/A",
    "sample_size": len(df_imputed),
    "conversion_rate_1": df_imputed_conversion_rate[1],
    "conversion_rate_2": df_imputed_conversion_rate[2],
    "t_statistic": t_stat_imputed,
    "p_value_t": p_value_t_imputed,
    "chi2_statistic": chi2_stat_imputed,
    "p_value_chi": p_value_chi_imputed
})

# 分層抽樣 (Stratified Sampling) A/B 測試結果. (Stratified Sampling A/B test results.)
for frac in frac_values:
    df_sampled = df_imputed.groupby("event_type", group_keys=False).apply(lambda x: x.sample(frac=frac, random_state=42))
    conversion_rate = df_sampled.groupby("variation")["converted"].mean()

    # 進行 T 檢定 (Perform a T test)
    group_1 = df_sampled[df_sampled["variation"] == 1]["converted"]
    group_2 = df_sampled[df_sampled["variation"] == 2]["converted"]
    t_stat, p_value_t = ttest_ind(group_2, group_1, equal_var=False)

    # 進行卡方檢定 (Chi-square test)
    conversion_table = pd.crosstab(df_sampled["variation"], df_sampled["converted"])
    chi2_stat, p_value_chi, _, _ = chi2_contingency(conversion_table)

    results.append({
        "method": "Stratified Sampling",
        "frac": frac,
        "sample_size": len(df_sampled),
        "conversion_rate_1": conversion_rate[1],
        "conversion_rate_2": conversion_rate[2],
        "t_statistic": t_stat,
        "p_value_t": p_value_t,
        "chi2_statistic": chi2_stat,
        "p_value_chi": p_value_chi
    })

# 轉為 DataFrame 方便查看 (Convert to DataFrame for easy viewing)
df_results = pd.DataFrame(results)
print("\nFraction Comparison Results:")
print(df_results)

# 可視化不同 frac 對轉化率影響. (Visualize the impact of different frac on conversion rate)
plt.figure(figsize=(12,5))
plt.plot(df_results[df_results["method"] == "Stratified Sampling"]["frac"], df_results[df_results["method"] == "Stratified Sampling"]["conversion_rate_1"], marker="o", linestyle="-", color="blue", label="Variation 1")
plt.plot(df_results[df_results["method"] == "Stratified Sampling"]["frac"], df_results[df_results["method"] == "Stratified Sampling"]["conversion_rate_2"], marker="o", linestyle="-", color="orange", label="Variation 2")
plt.title("Conversion Rate vs. Different Sampling Fractions")
plt.xlabel("Fraction (frac)")
plt.ylabel("Conversion Rate")
plt.legend()
plt.grid()
plt.show()

# 檢查 T 檢定的 P 值變化. (Examining changes in P values ​​for T tests)
plt.figure(figsize=(10,5))
plt.plot(df_results[df_results["method"] == "Stratified Sampling"]["frac"], df_results[df_results["method"] == "Stratified Sampling"]["p_value_t"], marker="o", linestyle="-", color="red", label="T-Test P-Value")
plt.axhline(y=0.05, color="black", linestyle="--", label="Significance Threshold (p=0.05)")
plt.title("T-Test P-Value vs. Different Sampling Fractions")
plt.xlabel("Fraction (frac)")
plt.ylabel("P-Value")
plt.legend()
plt.grid()
plt.show()
```
